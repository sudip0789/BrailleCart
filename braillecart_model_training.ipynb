```python
from ultralytics import YOLO
from pathlib import Path
from sklearn.metrics import precision_score, recall_score, average_precision_score

model_path = "/content/gdrive/My Drive/braillecart/detect/train/weights/best.pt"
data_path = "data.yaml"

model = YOLO(model_path)

# Validate the model on the validation dataset
results = model.val(data=data_path, save_json=True)

# Print metrics
metrics = results.results_dict

if metrics:
    print("Validation Metrics:")
    print(f"Precision: {metrics['metrics/precision(B)']:.4f}")
    print(f"Recall: {metrics['metrics/recall(B)']:.4f}")
    print(f"mAP@0.5: {metrics['metrics/mAP50(B)']:.4f}")
    print(f"mAP@0.5-0.95: {metrics['metrics/mAP50-95(B)']:.4f}")
else:
    print("Validation metrics could not be computed. Check your validation dataset and annotations.")
```
